```python
import numpy as np
import pandas as pd

n = 20
matematicas = np.around(np.random.normal(loc=85, scale=5, size=n), 2)
print(matematicas)

ingles = np.around(np.random.normal(loc=90, scale=5, size=n), 2)
print(ingles)

programacion = np.around(np.random.normal(loc=80, scale=5, size=n), 2)
print(programacion)
```

```python
# Crera dataframe
# Nobre de columnas
datos = pd.DataFrame({'matematicas': matematicas,
                   'ingles': ingles,
                   'programacion': programacion})
datos
```
